<h3 style = "color:#C23B23;">La régression linéaire</h3>

<p>L’algorithme de régression linéaire est un algorithme d’apprentissage supervisé c’est-à-dire qu’à partir de la variable cible ou de la variable à expliquer (Y), le modèle a pour but de faire une prédiction grâce à des variables dites explicatives (X) ou prédictives.</p>
<p> Exemple : Prédire la valeur d’une maison en fonction de sa superficie, sa localisation, la possibilité de parking ou non..
prédire le nombre d’utilisateurs et utilisatrices d’un service en ligne à un moment donné.</p>
<p> Un modèle de régression linéaire est un modèle de machine learning dont la variable cible (Y) est quantitative tandis que la variable X peut être quantitative ou qualitative.

L’objectif est de trouver une fonction dite de prédiction ou une fonction coût qui décrit la relation entre X et Y c’est-à-dire qu’à partir de valeurs connues de X, on arrive à donner une prédiction des valeurs de Y.</p>

<p> <center>La fonction recherchée est de la forme : Y= f(X) avec
    
    f(X) une fonction linéaire </p>

L'exercice 2 consiste à prédir si le pretest effectué sur les etudiants influencent sur ses résultats finals ou non : est ce que le pretest et le post test sont deux variables fortement corrélés ?? et est ce qu'on peut connaitre la resultat de post test à partir des resultats de prétest ??

<h3 style = "color:#4F9EC4;">Première méthode: </h3>

```python
import pyreadstat
test, meta = pyreadstat.read_sav('test_scores.sav')
test.head()
```

```python
test.info()
```

```python
import seaborn as sns
corr_matrix = test.corr()
sns.heatmap(corr_matrix, annot=True)
```

```python
corr_matrix['posttest'].sort_values(ascending=False)
```

on remarque que le target variable "posttest" est fortement corrélés avec les features : pretest, lunch et n_studient 

alors que le target n'est pas corrélé avec les autres features tel que : school_setting, school_tupe, teaching_méthode et gender ( relation d'indépendence entre ces features et le target)

==> Pour cela, on choisit de predir la valeur de target à partir des features fortement corrélés avec celui-ci seulement 

```python
# Prenons les 3 premiers attibuts les plus corrélés avec 'posttest'
from pandas.plotting import scatter_matrix
attributes = ['posttest', 'pretest', 'lunch', 'n_student']
scatter_matrix(frame=test[attributes], figsize=(12, 8))
plt.show()
```

```python
# ==> L'attribut le plus important dans la prédiction de `posttest` est `pretest`
test.plot(kind='scatter', x='pretest', y='posttest', figsize=(12,8), alpha=0.1)
plt.show()
```

Et pour mieux visualiser la dépendance et la corrélation entre "pretest" et post test, on va utiliser la bibliothèque matplotlib

```python
x_plt= test['pretest']
y_plt= test['posttest']
plt.xlabel('pre-test')
plt.ylabel('post-test')
plt.scatter(x_plt,y_plt)
```

```python
# Pour la séparation, on utilise la fonction train_test_split() de Scikit-Learn :
from sklearn.model_selection import train_test_split

X = test.drop({"gender", "school_setting", "teaching_method","school_type","posttest", "school", "classroom", "student_id"}, axis=1) 
y= test["posttest"].to_numpy()


X_train,X_test,y_train,y_test=train_test_split(X, y,test_size=0.2, shuffle=True,random_state=22)

print('X_train:', np.shape(X_train), 'X_test:', np.shape(X_test))
```

```python
X_train.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score 

#Création d'une instance (le modèle lin_reg) par le constructeur LinearRegression() 
lin_reg = LinearRegression()
# Apprentissage du modèle par la méthode fit() : Il s'agit d'une approche d'apprentissage
# supervisée puisqu'on utilise lin_reg.fit(X_train, y_train)
lin_reg.fit(X_train, y_train)

#Prédiction des les données d'apprentissage X_train par la méthode .predict()
y_pred = lin_reg.predict(X_train)
#Evaluation de la prédiction obtenue avec les deux métriques R2 et RMSE
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
r2 = r2_score(y_train, y_pred)
print("Training: R2=", r2, " et RMSE=", rmse)

#Prédiction sur les données de test X_test par la méthode .predict()
y_pred = lin_reg.predict(X_test)
#Evaluation de la prédiction obtenue avec les deux métriques R2 et RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("Testing: R2=", r2, " et RMSE=", rmse)
```

##### ***Le model obtenu donne des excellentes prédictions pour les données de train et les données de test, il s'agit d'un rightfit***

```python
from sklearn.pipeline import Pipeline
# K-fold cross-validation et GridSearchCV
pipelines = []
params = []
names = []


# ajouter GradientBoostingRegressor
pipelines.append(Pipeline([('clf', GradientBoostingRegressor())]))
params.append({'clf__n_estimators':[500,700,1000]})
names.append('GradientBoostingRegressor')

# ajouter RandomForestRegressor
pipelines.append(Pipeline([('clf', RandomForestRegressor())])) 
params.append({'clf__n_estimators': [50,100,200]})
names.append('RandomForestRegressor') 

# ajouter LinearRegression
pipelines.append(Pipeline([('clf', LinearRegression())])) 
params.append({'clf__normalize':[True]})
names.append('LinearRegression') 


#AdaBoostRegressor
pipelines.append(Pipeline([('clf', AdaBoostRegressor())])) ### AdaBoostRegressor
params.append({'clf__n_estimators':[500,1000]})
names.append('AdaBoostRegressor')

#ajouter KNeighborsRegressor
pipelines.append(Pipeline([('clf', KNeighborsRegressor())])) ### KNeighborsRegressor
params.append({'clf__n_neighbors':[1,2,3,4,5]})
names.append('KNeighborsRegressor')
```

```python
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score

def model(pipeline, parameters, name, X, y):    
    cv = KFold(n_splits=5, shuffle=True, random_state=32)
    grid_obj = GridSearchCV(estimator=pipeline, param_grid=parameters, cv=cv, scoring='r2', n_jobs=-1)
    grid_obj.fit(X,y)  
    print(name, 'R2:', grid_obj.best_score_)
    estimator = grid_obj.best_estimator_
    estimator.fit(X,y) # training sur tout training dataset
    return estimator 
estimators = []
for i in range(len(pipelines)):    
    estimators.append(model(pipelines[i], params[i], names[i], X_train, y_train))
```
